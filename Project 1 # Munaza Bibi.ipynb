```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('heart_disease_dataset.csv')
df.info()
print('first five rows')
print(df.head())
print('last_five_rows')
print(df.tail())
```

```python
Heart_Disease_counts = df.value_counts('Heart Disease')
print('Herat Disease Distribution')
print(Heart_Disease_counts)
mylabel = ['No Disease', 'Disease']
plt.pie(Heart_Disease_counts, labels = mylabel, autopct= '%1.1f%%', colors =('yellow','orange') )
plt.title('Heart Disease Distribution')
plt.show()
```

```python
props = df.groupby('Family History')['Heart Disease'].mean()
print(props)
plt.figure(figsize=(8, 6))
plt.bar(['Heart Disease without family history','Heart Disease with family history'] , props, color=['skyblue', 'hotpink'])
plt.xlabel('Family History of Heart Disease')
plt.ylabel('Proportion with Heart Disease')
plt.title('Proportion of Individuals with Heart Disease by Family History')
plt.ylim(0, 1)
plt.show()
```

```python
## Pattern and Corelations
```

**Pattern and** **Corelations**

**Cholesterol Levels**: High Cholestrol level Increase the Heart Diseases in higher rate.

**Blood Pressure and Sugar Blood**: are reltively high in average which are known as risk factor.

**Smoking and Alcohol Intake**: Heavy Alcohol ans Smoking create a strong corelation with Heart Disease.

**Exercise and Physical Activity**: Exercise on daily bases makes you protective against Heart Disease. More than 4 hours exercise are necessary per a week.


**Family History and Genetics**: Family History and Genetics are more corelate with Heart Disease.

**Obesity and Diabetes:** Obesity are also strongly correlate with Cholesterole and sugar level which are also risk factor.

**Stress and Heart Rate:** High stress levels and elevated heart rates are commonly observed in individuals with heart disease.

**Chest Pain Types:** Specific types of chest pain, particularly typical angina, are more common in individuals with heart disease.

**Exercise-Induced Angina:** The presence of exercise-induced angina is a significant risk factor and is more common in individuals with heart disease.

**Potential Risk Factors:**

**Modifiable:** Cholesterol levels, blood pressure, smoking, alcohol intake, physical activity, obesity, and stress levels are modifiable risk factors.

**Non-Modifiable:** Age, gender, and family history are non-modifiable risk factors but are crucial in assessing the overall risk.
